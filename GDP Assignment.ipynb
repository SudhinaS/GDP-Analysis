Since there are over 27 files containing sector-wise GSDP details for the various states and union territories, manually reading each can be tedious and error-prone.
```python
# Import library for reading the multiple files in one shot
import glob
```

Data Loading <br>
The datasets are already downloaded to the local machine as before. The files are picked up from the paths provided.
> For testing purpose, the file path needs to be changed accordingly.

```python
#Extract all the filepaths that match the given pattern i.e. csv files with name starting as 'NAD-'
file_names = glob.glob('Datasets/I-B/NAD-*.csv')
# Read all the matched files to a list
dfs = [pd.read_csv(f,encoding='unicode_escape') for f in file_names]
```

```python
# Extract  and add the name of state from filename as a new column for each file to avoid errors/confusion while merging them.
for dataframe, filename in zip(dfs, file_names):
    dataframe['State'] = filename.split('NAD-')[1].split('-GSVA')[0]
```

`ANALYSIS-4` <br>
`SECTOR_WISE CONTRIBUTION TO GSDP ACROSS THE STATES`

The sector-level(i.e. Primary, Secondary and Tertiary level) GSDP earnings data for the States for 2014-'15 is copied to a new dataframe.

```python
#Filter the sector-level and GSDP earnings for each state
per_sector_df = State_data.loc[State_data['Item'].isin(['Primary','Secondary','Tertiary','Gross State Domestic Product'])][['State','Item',"2014-15"]].sort_values(['State','Item',"2014-15"], ascending=[True,True,False])
# Rename the earnings column
per_sector_df.rename(columns={'2014-15':"Contribution('14-'15)"},inplace=True)
per_sector_df.head()
```

The data is pivoted so that the sectors show up as columns for ease of operations.

```python
Cols_ordered = ['State','Primary','Secondary','Tertiary','Gross State Domestic Product']
# Sum up the values for state using a pivot table
per_sector_df = pd.pivot_table(per_sector_df, index='State', columns='Item', values="Contribution('14-'15)")   #.reset_index()
per_sector_df.sort_values("Gross State Domestic Product", ascending=False, inplace=True)
per_sector_df.reset_index(inplace=True)
# Remove the name of the column headers 'idx'
del per_sector_df.columns.name
# Re-order the columns
per_sector_df = per_sector_df[Cols_ordered]
per_sector_df.head()
```

The percentage contribution per sector is computed for each state

```python
per_sector_df['% Contribution-Primary'] = (round(per_sector_df['Primary']/per_sector_df['Gross State Domestic Product'],2))*100
per_sector_df['% Contribution-Secondary'] = (round(per_sector_df['Secondary']/per_sector_df['Gross State Domestic Product'],2))*100
per_sector_df['% Contribution-Tertiary'] = (round(per_sector_df['Tertiary']/per_sector_df['Gross State Domestic Product'],3))*100
```

```python
# View the dataframe
per_sector_df
```

**Sector-wise GSDP Contribution Plot for the Indian states during 2014-2015**

```python
GDP_total = per_sector_df['Gross State Domestic Product']
sector_df = per_sector_df.iloc[:,0:4]
sector_percent_df = per_sector_df[per_sector_df.columns[1:4]].div(GDP_total, 0)*100

# plt.figure(figsize=(40,40))
plt.rcParams["figure.figsize"] = (15,15)
sector_df.plot(x = 'State', kind='barh',stacked=True, title = 'GDP Contribution per sector per state in 2014-15')

for sector in sector_percent_df:
    for i, (cum_sum, val, prcnt) in enumerate(zip(sector_df.iloc[:, 1:].cumsum(1)[sector], sector_df[sector], sector_percent_df[sector])): 
        if i <= 16:
            plt.text(cum_sum - val/2, i, "%.1f"%prcnt , va='center', ha='center')
```

`ANALYSIS-5`<br>

BINNING STATES INTO CATEGORIES

The states are categorized into four groups based on their per capita GDPs. The states with per capita between the 85th and the 100th percentile values of statewise per-capita GDP distribution are placed in C1. Those lying in the 50th to the 85th percentile bucket go to category C2, 20th-50th percentile ranged states to C3 and the rest to the last bucket C4. This is done to simplify the subsequent analysis.

```python
# Bin the states into four buckets according to the per_capita_gdp percentile region that they fall in
GDP_per_capita['Category'] = pd.qcut(GDP_per_capita["Per Capita GSDP during 2014-'15"], [0,0.20,0.5, 0.85, 1],['C_4','C_3','C_2','C_1'])
print(GDP_per_capita['Category'].value_counts())
GDP_per_capita.head()
```

The Category column from the per_capita_gdp dataframe is mapped to the state-wise dataframe based on the name of the state. So, the category of the state can be known from the state-level dataframe after mapping the new column.

```python
pd.options.mode.chained_assignment = None
# Map the category column back to the State dataframe
State_data.loc[:,'Category'] = State_data.loc[:,'State'].map(GDP_per_capita.set_index('State')['Category'])
State_data.head()
```

A new dataframe is created from the state data with GSDP information for all the sub-sectors for each category. The state-level dataframe contains data at state level, sector level, sub-sector level as well as sub-sub-sector level. The idea is to seclude the sub-sector level data and GSDP_per_state data from this. On examining the dataset, it can be seen that sub-sub-sector level are serialized below the sector-level data with a '.' character. This information is made use of to discard the sub-sub-sector data and the sub-sector rows are then selected by the 'Item' name. The columns are renamed to more meaningful names.

```python
# Create new df from state df by selecting categories column instead of state
Category_wise_data = State_data.loc[:,['Category','S.No.','Item','2014-15']]
# Select sub-sector columns alone. Expunge sector, sub-subsector and tax data columns
import re
Items_to_exclude = ['Primary','Secondary','Tertiary','TOTAL GSVA at basic prices','Taxes on Products','Subsidies on products',"Population ('00)",'Per Capita GSDP (Rs.)']
idxs = Category_wise_data.index[(~Category_wise_data['S.No.'].str.contains(re.escape(".")))&(~Category_wise_data['Item'].isin(Items_to_exclude))]
Category_wise_data = Category_wise_data.loc[idxs]
Category_wise_data.rename(columns={'2014-15':'Contibution (2014-15)','Item':'Sub-sector'},inplace=True)
Category_wise_data.head(30)
```

The data is summed up for each sub-sector at category level. The percentage contribution of each sub-sector is calculated as the ratio of sub-sector contribution to the total gdp contribution per category.

```python
# Groupby the category and Sub-sector column
Category_data_grouped = Category_wise_data.groupby(['Category','Sub-sector'], as_index=False).sum()
# Calculate sub-sector percentages as new column
Category_data_grouped['Percent_Contrib'] = ((Category_data_grouped['Contibution (2014-15)']/Category_data_grouped.groupby('Category')["Contibution (2014-15)"].transform('max'))*100).round(1)
Category_data_grouped.head(20)
```

Data for each category is divided into separate dfs. The cumulative percentage contribution is also calculated to find the top sub-sectors that add up to 80% of total GSDP.

#### CATEGORY 1

```python
Category_C1 = Category_data_grouped.loc[(Category_data_grouped['Category']=='C_1')&(Category_data_grouped['Sub-sector']!='Gross State Domestic Product')].sort_values('Percent_Contrib',ascending=False)
Category_C1.reset_index(drop=True,inplace=True)
Category_C1['Cumulative_Percent'] = Category_C1['Percent_Contrib'].cumsum().round(1)
Category_C1
```

```python
fig, ax = plt.subplots(figsize=(12,12), subplot_kw=dict(aspect="equal"))

label = Category_C1['Sub-sector']
percent = Category_C1['Percent_Contrib']
contribution = Category_C1['Contibution (2014-15)']

plt.rcParams.update({'font.size': 15})
    
patches, texts, autotexts = plt.pie(contribution,startangle=-40, autopct='', pctdistance=0.85)

#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

kw = dict(arrowprops=dict(arrowstyle="-"), zorder=0, va="center")

for i, p in enumerate(patches):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(label[i], xy=(x, y), xytext=(1.3*np.sign(x), 1.4*y),horizontalalignment=horizontalalignment, **kw)
    
for i, a in enumerate(autotexts):
    a.set_text("{}".format(percent[i]))

 
plt.title('Sub-Sector Contributions to GSDP - Category 1', fontsize=20, y=1.08)
# plt.tight_layout()
plt.show()
```

These changes are made and the 2014-2015 data for the states alone is filtered out for the analysis.

> NaN values are left untreated to avoid adding any bias to the data

```python
# Correct state names to match the previous dataframes
df_Dropout.at[[6,14,34],'Level of Education - State'] = ['Chhattisgarh','Jammu Kashmir','Uttarakhand']
df_Dropout
```

```python
# Correct the data error in the column name
df_Dropout.rename(columns={'Primary - 2014-2015.1':'Primary - 2015-2016'},inplace=True)
# Select only 2014-2015 year data for Primary, Upper-Primary and Secondary dropout
cols_to_select = list(filter(lambda x: (('2014-2015' in x) or ('State' in x)) and ('Senior' not in x), df_Dropout.columns))
rows_to_exclude = ['Chandigarh','Delhi','Puducherry','A & N Islands','Dadra & Nagar Haveli','Daman & Diu','Lakshadweep','All India']
df_Dropout =df_Dropout.loc[~(df_Dropout['Level of Education - State'].isin(rows_to_exclude)),cols_to_select].reset_index(drop=True)
df_Dropout
```

ANALYSIS

It is also noticed that states with two-worded names are recorded with an underscore separator in previous dataset while space separator is used here. So, the previous dataframe is formatted to match the current one. Then the per_capita_gdp of each state is mapped to the new dataframe from the old one.

```python
GDP_per_capita['State'] = GDP_per_capita['State'].str.replace('_',' ')
df_Dropout.loc[:,'GDP_per_capita'] = df_Dropout.loc[:,'Level of Education - State'].map(GDP_per_capita.set_index('State')["Per Capita GSDP during 2014-'15"])
df_Dropout
```

The correlation between drop-outs and per_capita_GDP is found.

```python
corr = df_Dropout.corr()
corr
```

The correlation is viewed in the form of a heatmap.

```python
plt.rcParams["figure.figsize"] = (10,8)
ax = sns.heatmap(corr, cmap="YlGnBu")
```

A pairplot is created with regression line added to give a better idea of the level of correlation.

```python
sns.set(font_scale=0.9)
pplt= sns.pairplot(df_Dropout[1:],kind="reg")
pplt.fig.suptitle("Correlation between School Dropout rate and Per Capita GDP during the year 2014-'15", y=1.05)
```

> **PLOT ANALYSIS**
* The school drop-out rate during primary, upper primary and secondary classes is negatively correlated with per capita GDP of states. i.e. As the school dropouts increases, the per capita GDP decreases.
* The negative GDP impact is more in case of upper-primary dropouts followed by secondary followed by a close primary dropout ratio.i.e. When the drop-outs happen in upper-primary classes, the fall in per capita GDP is maximum.
* Though it is an average correlation, the flippant impact is obvious. So is the reason. As the students drop out at lower classes, they are deprived of even basic formal education rendering them unequipped for any skilled labour or even an upgrade in an unskilled job. As the population of the state gets pooled with more dropouts, the availability of skilled labour drops which would gradually dry up that sub-sector/industry in the state. Not only does the non-productive jobs/sectors boom, this reduces sector diversification eating up job opportunities of skilled labourers who are then forced to be under-employed. This vicious cycle goes on retarding social as well as economic growth.
